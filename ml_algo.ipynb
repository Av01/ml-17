```python
import pandas as pd
mainList = []
corres_Dict = {'Front':1,'Left':2,'Right':3,'Rear':4}
Rows = txt.split('\n')
end = len(Rows)
for eachRow in Rows[1:end-1]:
#    print(eachRow)
    tokens = eachRow.split(',')
#    print(tokens)
    data = dict()
    data['ID'] = tokens[0]
    data['DetectedCamera'] = corres_Dict[tokens[1]]
    data['AngleOfSign'] = tokens[2]
    data['SignAspectRatio'] = tokens[3]
    data['SignWidth'] = tokens[4]
    data['SignHeight'] = tokens[5]
    data['SignFacing'] = corres_Dict[tokens[6]]
    mainList.append(data)

dataset = pd.DataFrame(mainList)
dataset.head()
```

```python
X = [[float(b) for b in list(w)] for w in dataset[['DetectedCamera','SignHeight','SignWidth','SignAspectRatio']].values]
y = [v for v in dataset['SignFacing'].values]
print(X[:10])
print(y[:10])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 0)
```
